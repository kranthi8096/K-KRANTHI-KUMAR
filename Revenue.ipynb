#### Group the DataFrame df by the 'Product' column andadd a new column 'Revenue Percentage' that shows the percentage of each product's revenue relative to the total revenue for that product.

```python
 df.groupby('Product')['Revenue'].transform(lambda x: (x / x.sum()) * 100)

# transform applies a function to each subgroup and preserves the original shape of the DataFrame.
# lambda x: (x / x.sum()) * 100 is an anonymous function (lambda) that takes a Series (x) as input.
```

## lambda:- 
Action in this code:
- Takes a Series (x) representing a subgroup's 'Revenue' column.
- Calculates the percentage of each revenue value relative to the total revenue within that subgroup.
- Returns a new Series with the calculated percentages.
